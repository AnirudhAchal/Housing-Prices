# Importing Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Setting path to test and training data
raw_data_path = os.path.join(os.path.pardir, 'Data', 'Raw_Data')
train_file_path = os.path.join(raw_data_path, 'train.csv')
test_file_path = os.path.join(raw_data_path, 'test.csv')
```

```python
# Loading test and training data
train_df = pd.read_csv(train_file_path, index_col = 'Id')
test_df = pd.read_csv(test_file_path, index_col = 'Id')
```
